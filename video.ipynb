```python
from google.cloud import videointelligence_v1 as videointelligence
import os

# Set the environment variable to point to your JSON key file
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "stoked-forest-447811-u4-ecf33505a9e7.json"

# Initialize the Video Intelligence client
client = videointelligence.VideoIntelligenceServiceClient()

# Specify the path to the video in Google Cloud Storage
video_uri = "gs://cloud-samples-data/video/animals.mp4"

# Define the features to analyze
features = [
    videointelligence.Feature.LABEL_DETECTION,
    videointelligence.Feature.SHOT_CHANGE_DETECTION,
]

# Start the video analysis request
operation = client.annotate_video(
    request={"features": features, "input_uri": video_uri}
)

print("Processing video for label and shot change detection...")
result = operation.result(timeout=90)  # Wait for the operation to complete

# Process the results
print("\nFinished processing.\n")

# Label Detection Results
for annotation in result.annotation_results[0].segment_label_annotations:
    print(f"Label: {annotation.entity.description}")
    for segment in annotation.segments:
        start_time = segment.segment.start_time_offset.total_seconds()
        end_time = segment.segment.end_time_offset.total_seconds()
        confidence = segment.confidence
        print(
            f"\tSegment: {start_time}s to {end_time}s, Confidence: {confidence:.2f}"
        )

# Shot Change Detection Results
for annotation in result.annotation_results[0].shot_annotations:
    start_time = annotation.start_time_offset.total_seconds()
    end_time = annotation.end_time_offset.total_seconds()
    print(f"Shot change detected at: {start_time}s to {end_time}s")
```
